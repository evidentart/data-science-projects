# Creating Visualizations for Data Analysis

### Task 1.1: Create a line chart with pandas to visualize how automobile sales change and fluctuate across different years

```python
plt.figure(figsize=(10, 6))

# Group by Year and calculate average automobile sales
df_line = df.groupby('Year')['Automobile_Sales'].mean()

# Plot the line chart
df_line.plot(kind='line', marker='o', label='Avg Automobile Sales')

# X-axis ticks for all years from 1980 to 2023
plt.xticks(list(range(1980, 2024)), rotation=75)

# Labels and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

# Annotate two recession periods
plt.text(1982, 650, '1981-82 Recession')
plt.text(2009, 500, '2008-09 Recession')

plt.legend()
plt.grid(True)
plt.show()
```

#### Inference: 
The chart shows that automobile sales dropped significantly during major recessions (1981–82 and 2008–09), with sharp declines visible around those periods. Sales generally recovered afterward, but fluctuations remained, showing how recessions negatively impacted the auto industry.

### Task 1.2: Create line charts for each vehicle type to visualize and compare how sales trends vary across categories, especially during recession periods

```python
df_rec = df[df['Recession'] == 1]

df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Normalize sales per vehicle type
df_Mline['Normalized_Sales'] = df_Mline.groupby('Vehicle_Type')['Automobile_Sales'].transform(lambda x: x / x.mean())

df_Mline.set_index('Year', inplace=True)

plt.figure(figsize=(12, 8))

for vehicle_type in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vehicle_type]
    plt.plot(data.index, data['Normalized_Sales'], label=vehicle_type, marker='o', linewidth=2)

# Vertical lines for recession years
recession_years = sorted(df_rec['Year'].unique())
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Normalized Sales")
plt.xlabel("Year")
plt.title("Normalized Automobile Sales by Vehicle Type During Recession")
plt.tight_layout()
plt.show()
```

#### Inference: 
Sports cars and supermini cars demonstrate resilience or growth during recession periods. Medium family cars and, to a lesser extent, small family cars show more sensitivity to economic changes, with less consistent trends.The upward trend in sports vehicles sales indicates the stability of the luxury market, even during economic downturns.

### Task 1.3: Compare sales trends per vehicle type between recession and non-recession periods using Seaborn

```python
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(x=new_df['Recession'], height=new_df['Automobile_Sales'], color=['#1f77b4', '#ff7f0e'])
plt.xlabel('Period')
plt.ylabel('Average Automobile Sales')
plt.title('Average Automobile Sales during Recession and Non-Recession')
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.show()
```

#### Inference: 
This bar chart shows that average automobile sales during recessions are much lower compared to non-recession periods. Sales drop sharply in recessions, highlighting the strong negative impact of economic downturns on the auto industry.

### Compare automobile sales across vehicle types for recession vs non-recession periods

```python
# Group by Recession and Vehicle_Type, calculate average sales
dd = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

plt.figure(figsize=(12, 6))

# Grouped barplot
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=dd)

# X-axis labels for readability
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])

# Labels and title
plt.xlabel('Recession Period')
plt.ylabel('Average Automobile Sales')
plt.title('Vehicle-Wise Sales during Recession and Non-Recession')

plt.legend(title='Vehicle Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside
plt.tight_layout()
plt.show()
```

#### Inference: 
The plot shows a significant decline in overall automobile sales during recession periods. Executive and sports cars are the most affected, experiencing the largest drop in sales.

### Task 1.4: Compare GDP trends during recession and non-recession periods using subplots

```python
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

fig = plt.figure(figsize=(14, 6))

ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

sns.lineplot(x='Year', y='GDP', data=rec_data, ax=ax0, marker='o', color='#1f77b4')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during Recession Period')

sns.lineplot(x='Year', y='GDP', data=non_rec_data, ax=ax1, marker='o', color='#1f77b4')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during Non-Recession Period')

plt.tight_layout()
plt.show()
```

#### Inference
From this plot, it is evident that during recession, the GDP of the country was in a low range, might have afected the overall sales of the company

### Task 1.5: Analyze the impact of seasonality on automobile sales using a bubble plot

```python
# Filter non-recession years
non_rec_data = df[df['Recession'] == 0]

plt.figure(figsize=(12, 6))

# Bubble plot with Seasonality_Weight as bubble size
sns.scatterplot(
    data=non_rec_data, 
    x='Month', 
    y='Automobile_Sales', 
    size='Seasonality_Weight', 
    legend='full',              
    sizes=(50, 500),             
    color='#1f77b4'              
)

# Labels and title
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales')
plt.xticks(range(1, 13))  # ensure all months 1-12 are visible
plt.show()
```

#### Inference
From this plot, it is evident that seasonality has not affected on the overall sales. However, there is a drastic raise in sales in the month of April

 ### Task 1.6: Explore the relationship between average vehicle price and sales volume during recession periods

```python
# Filter recession period data
recession_data = df[df['Recession'] == 1]

plt.figure(figsize=(8, 6))

# Scatter plot: Price vs Automobile Sales
plt.scatter(
    recession_data['Price'], 
    recession_data['Automobile_Sales'], 
    color='#1f77b4', 
    alpha=0.7
)

plt.xlabel('Vehicle Price')
plt.ylabel('Automobile Sales')
plt.title('Vehicle Price vs Automobile Sales during Recessions')
plt.show()
```

### Task 1.7: Visualize the share of advertising expenditure during recession and non-recession periods

```python
# Filter data
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Sum advertising expenditure
RAtotal = recession_data['Advertising_Expenditure'].sum()
NRAtotal = non_recession_data['Advertising_Expenditure'].sum()

# Pie chart
plt.figure(figsize=(8, 6))
labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]

plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#ff7f0e', '#1f77b4'],  # visually appealing colors
    shadow=True
)

plt.title('Advertising Expenditure of XYZ Automotives during Recession and Non-Recession Periods')
plt.show()
```

#### Inference 
The pie chart shows that XYZ Automotives spends significantly more on advertising during non-recession periods than during recessions. This indicates a reduction in marketing expenditure when economic conditions are tougher.

### Task 1.8: Visualize advertising expenditure by vehicle type during recession periods

```python
# Filter recession data
recession_data = df[df['Recession'] == 1]

# Sum advertising expenditure by vehicle type
vehicle_expenditure = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Pie chart with bold, distinct colors
plt.figure(figsize=(8, 8))
labels = vehicle_expenditure.index
sizes = vehicle_expenditure.values

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.tab20.colors,  # bold, highly distinct colors
    shadow=True
)

plt.title('Share of Each Vehicle Type in Total Expenditure during Recessions')
plt.show()
```

#### Inference 
During recessions, advertising was primarily focused on lower-priced vehicles, reflecting a strategic shift to target more price-sensitive customers.

 ### Task 1.9: Analyze the effect of unemployment rate on vehicle type sales during recession

```python
# Filter recession period data
recession_data = df[df['Recession'] == 1]

plt.figure(figsize=(10, 6))

sns.lineplot(
    data=recession_data,
    x='unemployment_rate',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    style='Vehicle_Type',
    markers=True,
    dashes=False,
    err_style=None
)

plt.ylim(0, 850)
plt.xlabel('Unemployment Rate')
plt.ylabel('Automobile Sales')
plt.title('Effect of Unemployment Rate on Vehicle Type Sales During Recession')

# Move legend to lower left and make it smaller
plt.legend(title='Vehicle Type', loc='lower left', fontsize=9, title_fontsize=10)
plt.grid(True)
plt.show()
```

#### Inference 
During recession periods, sales of lower-priced vehicles such as supermini cars and small family cars decrease as unemployment rises, showing higher sensitivity to economic downturns. Medium mini cars remain relatively stable, indicating less impact from changes in unemployment.

### Task 1.10: Map the regions with highest sales during recession

```python
# Filter recession data
df_rec = df[df['Recession'] == 1]
sales_by_state = df_rec.groupby('City')['Automobile_Sales'].sum().reset_index()

# Approximate coordinates for the states
coords = {
    'Georgia': [32.1656, -82.9001],
    'New York': [43.0000, -75.0000],
    'Illinois': [40.0000, -89.0000],
    'California': [37.2500, -119.7500]
}

# Create the map with smaller size
m = folium.Map(location=[37.8, -96], zoom_start=4, width=600, height=400)

# Add circle markers for each state
for _, row in sales_by_state.iterrows():
    state = row['City']
    sales = row['Automobile_Sales']
    folium.CircleMarker(
        location=coords[state],
        radius=sales / 5000,  # scale bubble size
        color='blue',
        fill=True,
        fill_opacity=0.6,
        popup=f"{state}: {sales} sales"
    ).add_to(m)

m
```
